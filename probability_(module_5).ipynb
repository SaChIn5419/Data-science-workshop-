```python
import random

# Tossing a coin 10,000 times
def coin_toss_simulation():
    heads, tails = 0, 0
    for _ in range(10000):
        toss = random.choice(["Heads", "Tails"])
        if toss == "Heads":
            heads += 1
        else:
            tails += 1
    print(f"Probability of Heads: {heads / 10000}")
    print(f"Probability of Tails: {tails / 10000}")

coin_toss_simulation()
```

```python
# Rolling two dice to compute the probability of getting a sum of 7
def dice_roll_simulation():
    count_sum_7 = 0
    trials = 10000

    for _ in range(trials):
        dice1 = random.randint(1, 6)
        dice2 = random.randint(1, 6)
        if dice1 + dice2 == 7:
            count_sum_7 += 1

    print(f"Probability of getting a sum of 7: {count_sum_7 / trials}")

dice_roll_simulation()
```

```python
# Probability of getting at least one "6" in 10 rolls
def at_least_one_six():
    trials = 10000
    success = 0

    for _ in range(trials):
        if any(random.randint(1, 6) == 6 for _ in range(10)):
            success += 1

    print(f"Probability of at least one '6' in 10 rolls: {success / trials}")

at_least_one_six()
```

```python
# Simulating 1000 draws from a bag
def ball_draw_simulation():
    colors = ["Red"] * 5 + ["Green"] * 7 + ["Blue"] * 8
    draws = 1000
    blue_red_count = 0
    blue_count = 0

    for _ in range(draws):
        ball1 = random.choice(colors)
        ball2 = random.choice(colors)
        if ball1 == "Blue":
            blue_count += 1
            if ball2 == "Red":
                blue_red_count += 1

    conditional_prob = blue_red_count / blue_count if blue_count else 0
    print(f"Probability of Red given the previous was Blue: {conditional_prob}")

ball_draw_simulation()
```

```python
import numpy as np

# Generating a sample and calculating statistics
def discrete_random_variable():
    probabilities = [0.25, 0.35, 0.4]
    outcomes = [1, 2, 3]
    sample = np.random.choice(outcomes, size=1000, p=probabilities)

    mean = np.mean(sample)
    variance = np.var(sample)
    std_dev = np.std(sample)

    print(f"Empirical Mean: {mean}")
    print(f"Variance: {variance}")
    print(f"Standard Deviation: {std_dev}")

discrete_random_variable()
```

```python
import matplotlib.pyplot as plt

# Simulating exponential distribution
def exponential_distribution():
    data = np.random.exponential(scale=5, size=2000)

    plt.hist(data, bins=30, density=True, alpha=0.6, color='g', label="Histogram")

    # Plotting the PDF
    x = np.linspace(0, 30, 1000)
    pdf = (1 / 5) * np.exp(-x / 5)
    plt.plot(x, pdf, color='r', label="PDF")

    plt.title("Exponential Distribution")
    plt.legend()
    plt.show()

exponential_distribution()
```

```python
# Central Limit Theorem
def central_limit_theorem():
    # Uniform distribution
    data = np.random.uniform(0, 1, 10000)

    # Sampling and calculating means
    sample_means = [np.mean(np.random.choice(data, size=30)) for _ in range(1000)]

    # Plotting
    plt.figure(figsize=(12, 6))

    plt.subplot(1, 2, 1)
    plt.hist(data, bins=30, color='b', alpha=0.6, label="Uniform Distribution")
    plt.title("Uniform Distribution")
    plt.legend()

    plt.subplot(1, 2, 2)
    plt.hist(sample_means, bins=30, color='g', alpha=0.6, label="Sample Means")
    plt.title("Sample Means Distribution (CLT)")
    plt.legend()

    plt.tight_layout()
    plt.show()

central_limit_theorem()
```
