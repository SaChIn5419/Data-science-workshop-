```python
import numpy as np

# 1. Create a 1D NumPy array
array_1d = np.arange(1, 21)  # Integers from 1 to 20

# a. Calculate sum, mean, median, and standard deviation
array_sum = np.sum(array_1d)
array_mean = np.mean(array_1d)
array_median = np.median(array_1d)
array_std = np.std(array_1d)

print("Sum:", array_sum)
print("Mean:", array_mean)
print("Median:", array_median)
print("Standard Deviation:", array_std)

# b. Find indices of elements greater than 10
indices_gt_10 = np.where(array_1d > 10)
print("Indices of elements greater than 10:", indices_gt_10)
```

```python
# 2. Create a 2D NumPy array
array_2d = np.arange(1, 17).reshape(4, 4)

# a. Print the array
print("2D Array:\n", array_2d)

# b. Find the transpose
transpose_array = array_2d.T
print("Transpose of Array:\n", transpose_array)

# c. Calculate row-wise and column-wise sums
row_sums = np.sum(array_2d, axis=1)
col_sums = np.sum(array_2d, axis=0)

print("Row-wise sums:", row_sums)
print("Column-wise sums:", col_sums)
```

```python
# 3. Create two 3x3 random arrays
array1 = np.random.randint(1, 21, size=(3, 3))
array2 = np.random.randint(1, 21, size=(3, 3))

# a. Perform element-wise operations
addition = array1 + array2
subtraction = array1 - array2
multiplication = array1 * array2

print("Array 1:\n", array1)
print("Array 2:\n", array2)
print("Addition:\n", addition)
print("Subtraction:\n", subtraction)
print("Multiplication:\n", multiplication)

# b. Compute the dot product
dot_product = np.dot(array1, array2)
print("Dot Product:\n", dot_product)
```

```python
# 4. Reshape a 1D array into 3x4 and slice
array_reshaped = np.arange(1, 13).reshape(3, 4)
sliced_array = array_reshaped[:2, -2:]

print("Reshaped Array:\n", array_reshaped)
print("Sliced Array (First 2 Rows, Last 2 Columns):\n", sliced_array)
```

```python
import pandas as pd

# Create a DataFrame
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'Diana', 'Eve'],
    'Age': [24, 27, 22, 32, 29],
    'Department': ['HR', 'Finance', 'IT', 'Marketing', 'HR'],
    'Salary': [45000, 54000, 50000, 62000, 47000]
}
df = pd.DataFrame(data)

# a. Print the first five rows
print("First Five Rows:\n", df)

# b. Summary statistics
print("\nSummary Statistics:\n", df[['Age', 'Salary']].describe())

# c. Average salary in HR department
hr_avg_salary = df[df['Department'] == 'HR']['Salary'].mean()
print("\nAverage Salary in HR Department:", hr_avg_salary)
```

```python
# 2. Add a new column 'Bonus' (10% of salary)
df['Bonus'] = df['Salary'] * 0.10
print("\nDataFrame with Bonus:\n", df)
```

```python
# 3. Filter employees aged between 25 and 30
filtered_df = df[(df['Age'] >= 25) & (df['Age'] <= 30)]
print("\nEmployees Aged Between 25 and 30:\n", filtered_df)
```

```python
# 4. Group by 'Department' and calculate average salary
avg_salary_by_dept = df.groupby('Department')['Salary'].mean()
print("\nAverage Salary by Department:\n", avg_salary_by_dept)
```

```python
# 5. Sort by 'Salary' in ascending order
sorted_df = df.sort_values(by='Salary')
print("\nSorted DataFrame by Salary:\n", sorted_df)

# Save to CSV
sorted_df.to_csv('sorted_employees.csv', index=False)
```

```python
import matplotlib.pyplot as plt

# Data for line plot
x = [1, 2, 3, 4, 5]
y = [10, 15, 25, 30, 50]

# a. Plot the data
plt.plot(x, y, marker='o')
plt.title("Simple Line Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.grid(True)
plt.show()
```

```python
# Data for bar graph
students = ['John', 'Jane', 'Alice', 'Bob']
marks = [75, 85, 60, 90]

# a. Plot the bar graph
plt.bar(students, marks, color=['blue', 'orange', 'green', 'red'])
plt.title("Marks Scored by Students")
plt.xlabel("Students")
plt.ylabel("Marks")
plt.show()
```

```python
# Data for pie chart
regions = ['North America', 'Europe', 'Asia', 'Others']
revenue = [45, 25, 20, 10]

# a. Create pie chart
explode = [0.1 if r == max(revenue) else 0 for r in revenue]  # Highlight max revenue
plt.pie(revenue, labels=regions, autopct='%1.1f%%', explode=explode, shadow=True)
plt.title("Revenue Distribution by Region")
plt.show()
```
