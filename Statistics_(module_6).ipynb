```python
# Step 1: Create a dataset with outliers
import numpy as np
import matplotlib.pyplot as plt
from statistics import mean, median, mode

# Dataset with and without outliers
data = [10, 12, 12, 13, 12, 11, 10, 300]  # 300 is the outlier
data_no_outliers = [10, 12, 12, 13, 12, 11, 10]

# Step 2: Calculate mean, median, and mode
print("With Outliers:")
print("Mean:", mean(data))
print("Median:", median(data))
print("Mode:", mode(data))

print("\nWithout Outliers:")
print("Mean:", mean(data_no_outliers))
print("Median:", median(data_no_outliers))
print("Mode:", mode(data_no_outliers))

# Inference:
# - The mean is highly sensitive to outliers, as seen when 300 skews the average upward.
# - The median is more robust, showing minimal change even with the outlier.
# - The mode remains unaffected since it relies on the most frequent values.
```

```python
# Step 1: Use a dataset of test scores
import seaborn as sns

test_scores = [55, 60, 65, 70, 75, 80, 85, 90, 95, 100, 100, 85, 75]

# Step 2: Calculate mean, median, and mode
print("Mean:", mean(test_scores))
print("Median:", median(test_scores))
print("Mode:", mode(test_scores))

# Step 3: Plot histogram
plt.figure(figsize=(8, 5))
sns.histplot(test_scores, bins=10, kde=True, color="skyblue", edgecolor="black")
plt.axvline(mean(test_scores), color='red', label='Mean', linestyle='--') #optional
plt.axvline(median(test_scores), color='green', label='Median', linestyle='--')#optional
plt.axvline(mode(test_scores), color='blue', label='Mode', linestyle='--')#optional
plt.legend()
plt.title("Histogram of Test Scores with Measures of Central Tendency")
plt.show()
# Inference:
# - The histogram shows the distribution of test scores.
# - The mean, median, and mode are overlaid as vertical lines to observe their positions.
# - This visualization helps identify how the scores are distributed (symmetrical or skewed).
```

```python
# Step 1: Calculate range, IQR, variance, and standard deviation
import scipy.stats as stats

data = [15, 18, 21, 21, 24, 30, 34, 40, 45]

range_value = max(data) - min(data)
iqr = stats.iqr(data)
variance = np.var(data, ddof=1)  # Sample variance
std_dev = np.std(data, ddof=1)  # Sample standard deviation

print("Range:", range_value)
print("IQR:", iqr)
print("Variance:", variance)
print("Standard Deviation:", std_dev)

# Step 2: Visualize using boxplot
plt.figure(figsize=(6, 4))
sns.boxplot(data=data, color="lightblue")
plt.title("Boxplot of Data")
plt.show()
# Inference:
# - The range, IQR, variance, and standard deviation quantify how spread out the data is.
# - The boxplot highlights the IQR and potential outliers.
# - High variance or standard deviation indicates that data points deviate significantly from the mean.
```

```python
# Step 1: Create datasets for scores before and after the program
from scipy.stats import ttest_rel

before = [50, 55, 60, 65, 70]
after = [60, 65, 70, 75, 80]

# Step 2: Perform paired t-test
t_stat, p_value = ttest_rel(before, after)
print("T-statistic:", t_stat)
print("P-value:", p_value)

# Step 3: Interpret p-value
if p_value < 0.05:
    print("The study program significantly improved the scores.")
else:
    print("No significant improvement in scores due to the program.")
# Inference:
# - A paired t-test compares two related samples to determine if there is a statistically significant difference.
# - A low p-value (<0.05) indicates the program was effective.
```

```python
# Step 1: Type I and Type II error simulation
import random

# Synthetic data for population mean = 50
population = [random.gauss(50, 5) for _ in range(1000)]

# Type I Error (False positive: Rejecting a true null)
sample1 = random.sample(population, 30)
t_stat1, p_value1 = stats.ttest_1samp(sample1, 50)
print("Type I Error Scenario:")
print("P-value:", p_value1)

# Type II Error (False negative: Failing to reject a false null)
population_shifted = [x + 10 for x in population]
sample2 = random.sample(population, 30)
t_stat2, p_value2 = stats.ttest_1samp(sample2, 50)
print("\nType II Error Scenario:")
print("P-value:", p_value2)
# Inference:
# - Type I error: A significant result when the null hypothesis is true.
# - Type II error: Failure to detect a significant effect when one exists.
```

```python
# Step 1: Fit a simple linear regression model
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

# Synthetic data
X = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)  # Years of experience
y = np.array([30, 35, 50, 65, 80])  # Salaries

# Train the model
model = LinearRegression()
model.fit(X, y)

# Predict for 5 years
predicted_salary = model.predict([[5]])
print("Predicted Salary for 5 years of experience:", predicted_salary[0])

# Step 2: Visualize regression line
plt.scatter(X, y, color='blue', label='Data')
plt.plot(X, model.predict(X), color='red', label='Regression Line')
plt.xlabel("Years of Experience")
plt.ylabel("Salary")
plt.legend()
plt.title("Simple Linear Regression")
plt.show()
# Inference:
# - The regression line models the relationship between years of experience and salary.
# - Predictions, such as salary for 5 years of experience, use this line.
```

```python
# Step 1: Generate quadratic dataset with noise
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import make_pipeline

np.random.seed(42)
X = np.linspace(0, 10, 100).reshape(-1, 1)
y = 3 * X**2 - 7 * X + 5 + np.random.normal(0, 5, X.shape)

# Fit models
linear_model = LinearRegression()
linear_model.fit(X, y)

poly_model = make_pipeline(PolynomialFeatures(10), LinearRegression())
poly_model.fit(X, y)

# Plot
plt.scatter(X, y, label='Data', alpha=0.6)
plt.plot(X, linear_model.predict(X), label='Underfitting (Linear)', color='red')
plt.plot(X, poly_model.predict(X), label='Overfitting (High-Degree Poly)', color='green')
plt.legend()
plt.title("Underfitting vs Overfitting")
plt.show()
# Inference:
# - The linear model (underfitting) cannot capture the quadratic relationship, leading to high error.
# - The high-degree polynomial model (overfitting) captures noise in the data, leading to poor generalization.
```

```python
# Step 1: Generate dataset
X = np.linspace(0, 10, 100).reshape(-1, 1)
y = 2 * X + 1 + np.random.normal(0, 1, X.shape)

# Models
from sklearn.linear_model import Lasso

linear = LinearRegression()
poly = make_pipeline(PolynomialFeatures(3), LinearRegression())
lasso = Lasso(alpha=0.1)

for model, name in [(linear, 'Linear'), (poly, 'Polynomial'), (lasso, 'Lasso')]:
    model.fit(X, y)
    y_pred = model.predict(X)
    mse = np.mean((y - y_pred)**2)
    print(f"{name} Regression MSE:", mse)
# Inference:
# - Linear regression is simple but may not capture complex relationships.
# - Polynomial regression fits non-linear patterns but may overfit with high-degree terms.
# - Lasso regression prevents overfitting by penalizing large coefficients.
```

```python
# Step 1: Correlation and scatterplot
from scipy.stats import pearsonr

X = np.array([1, 2, 3, 4, 5])
y = np.array([2, 4, 5, 4, 5])

# Correlation coefficient
correlation, _ = pearsonr(X, y)
print("Pearson Correlation Coefficient:", correlation)

# Fit regression
reg_model = LinearRegression()
reg_model.fit(X.reshape(-1, 1), y)

# Scatterplot with regression line
plt.scatter(X, y, color='blue', label='Data')
plt.plot(X, reg_model.predict(X.reshape(-1, 1)), color='red', label='Regression Line')
plt.legend()
plt.title("Scatterplot with Regression Line")
plt.show()
# Inference:
# - The Pearson correlation coefficient quantifies the strength of the linear relationship.
# - The regression line shows the best-fit relationship between the variables.
```
